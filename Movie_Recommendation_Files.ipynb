```python
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
  # Converts genre JSON strings to a list of genre names
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast']
```

```python
import ast
def convert3(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter==3:
      L.append(i['name'])
      counter = counter+1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
import ast
def fetch_director(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if i ['job']:
      L.append(i['name'])
  return L
  # Function to extract the director from the crew data
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'][0]
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: "".join(x))
# Removing unnecessary columns and handling null values
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
# Converting text data into numerical vectors using CountVectorizer
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Calculating cosine similarity between movie vectors
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(similarity[0])), reverse = True,key = lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
